# The Convex Hull Problem

The *convex hull* of a finite set of planar points is the smallest convex polygon that contains all the points. If the points are nails in a board, then the convex hull is what a rubberband around the nails would snap to. In the diagram below, the green nails define the convex hull of the set of nails. (A *convex polygon* is defined as one where a line segment connecting any two vertexes lies within the polygon (or on the edge).) 

<img src="http://www.personal.kent.edu/~rmuhamma/Compgeometry/MyCG/Gifs-CompGeometry/ch2.gif">

In this exercise we develop two algorithms to find the convex hull: first an incremental algorithm that updates the hull each time a new point is added to the set; then a batch algorithm that is given a set of points all at once and computes the hull once-and-for-all.

## An Incremental Convex Hull Finder

We will define the `IncrementalHull` class, which works as follows: First you create an object of the class with `h = IncrementalHull()`; then you add points, one at a time, with `h.add(point)`; then at any time you can retrieve the points in the hull with `h.hull`.  The result will be a list, in sorted (counterclockwise) order, of the points in the hull.  

The key idea of the algorithm is that if we have an actual convex hull and we follow the perimeter in counterclockwise order, then at every vertex we will be making a left turn.  If there were ever a vertex that forms a right turn, like the vertex *p<sub>2</sub>* below, that would mean the vertex is actually in the interior of the convex hull, so the vertex should be removed from the hull.

<img src="http://users.cis.fiu.edu/~giri/teach/UoM/7713/f98/stara/image6.gif">

For the incremental algorithm, we assume that we have a proper convex hull, and then we are presented with a new point.  We have to first figure out where the point should go in counterclockwise order, and then check to see if it is the elbow of a left turn.  If it is not, then the point is discarded; removed from the hull.  If it is a left turn, then it is in.  However, if the point is kept, it means that it pushes the boundary of the hull outward; this might cause other neighboring points to now be in the interior, and thus those points will need to be removed from the hull.  We need to search in both directions from the new point, removing a neighboring point if it does not form a left turn. If we remove a neighbor, we then check the next neighbor.  Once the neighbor forms a left turn we are done; we don't have to look (in that direction) any further.

That's all there is to the basic idea of the algorithm, but there are two edge cases to worry about:  

* *Colinear points:* if three or more points are colinear, we should keep only the two "outside" ones.  The rationale for this is that we want the convex hull to be the minimal possible set of points.

* *Degenerate polygons*: You can't have a polygon with  fewer than 3 vertexes, so what is the convex hull of 0 or 1 or 2 points?  We will define the convex hull of a set of two or fewer points to be the points themselves, even though they don't form a "real" polygon.

There are also some implementation choices to make:

* *Counterclockwise ordering:* We need a way to keep the points currently on the hull sorted, in counterclockwise order, and to insert a new point into the right place in the order.  To do that, we will use a separate class, `CircularSorted`.  It maintains a sorted list of elements, and, because it is a subclass of list, provides the same interface a list does.  In addition, it provides `before` and `after` methods, so that `before(P)` returns the point that comes before `P` in counterclockwise order.

* *Anchor point*: We need some way of telling the `CircularSorted` list what to sort by.  We will do that by associating with each point in the list the angle it makes with a central *anchor point*.  We will make sure that the anchor is in the interior of the hull, and will make the `key` function for the `CircularSorted` list be a function that computes the angle to this anchor point. The function `angle_from(A)` returns a function that computes this angle. (The angle ranges from -&pi; to +&pi;.  If point `A` is the origin, then the angle from `A` is 0 for points on the positive x-axis; &pi;/2 for points on the positive y-axis; &pi; for points on the negative x-axis; and -&pi;/2 for points on the negative y-axis.)

* *Points not on hull:* While the main job of `IncrementalHull` is to maintain the hull, we will also keep a list of all the points that were presented (by `add(point)`), in the order in which they were presented.  We'll keep them in the `.history` attribute.



```python
class IncrementalHull(object):
    "Incrementally maintain a convex hull (in .hull) as you .add(point)."

    def __init__(self):
        "Construct a new Hull, initially with no points."
        self.anchor = None            # The anchor/central point
        self.hull = CircularSorted()  # A sequence of points in sorted order (by angle from anchor)
        self.history = []             # A list of points in the order presented to add(); some on hull, some not
                
    def add(self, point):
        "Incrementally update hull by adding point to history and maybe to hull; maybe remove other points from hull."
        self.history.append(point)
        hull, before, after = self.hull, self.hull.before, self.hull.after # abbreviations, for convenience
        if len(hull) < 3: 
            # First 3 points: set the hull to them, with the average of the points as the new anchor
            points = hull + [point]
            # Special case: if 3 colinear points, keep only the two extremes
            if len(points) == 3 and turn(points[0], points[1], points[2]) == 'straight':
                points = [min(points), max(points)] 
            self.anchor = centroid(points)
            hull.set(points, key=angle_from(self.anchor))
        else: # General case: add point but then remove it if it doesn't make a left turn
            hull.add(point) 
            if turn(before(point), point, after(point)) == 'left': 
                # point is in exterior; see if it removes neighboring points
                while turn(point, after(point), after(after(point))) != 'left': 
                    hull.remove(after(point))
                while turn(before(before(point)), before(point), point) != 'left': 
                    hull.remove(before(point))
            else: # point is in interior, so remove it.
                hull.remove(point)
```

Here are some definitions: for the Point class, and some helper functions for Points, angles, turns, and averages:

```python
Point = namedtuple('Point', 'x, y') 

def Points(*coordinate_pairs):
    "Make a list of points: Convert (x, y) pairs into Point(x, y) objects."
    return [Point(x, y) for (x, y) in coordinate_pairs]

def angle_from(A):
    """Return a function that takes a point, B, as input and returns the angle from A to B.
    This function is suitable as the key function in a call to list.sort(key=angle_from(A))."""
    ## The atan2 angle value is 0 for a point B 
    return lambda B: atan2(B.y-A.y, B.x-A.x)

def turn(A, B, C):
    "Is the turn from a -> b -> c a left, right, or straight?"
    val = ((B.x - A.x) * (C.y - A.y)  -  (B.y - A.y) * (C.x - A.x))
    return 'left' if (val > 0) else 'right' if (val < 0) else 'straight'

def centroid(points):
    "The average of a collection of points (the center of gravity of a polygon)."
    return Point(average(p.x for p in points), average(p.y for p in points))

def average(numbers):
    "The average or mean of these numbers."
    numbers = list(numbers)
    return sum(numbers) / float(len(numbers))
```

Now for the CircularSorted class.  It is a subclass of `list`, so in one sense it is just a list of points.  But it is more than a normal `list` in that it has a `add` method that adds a new item and puts it into sorted order; it has a `key` attribute that can be `set` to a function that determines the sort order; and it has `before` and `after` methods that give you the next item in the sorted ordering.  These wrap around: the `after` of the last item is the first item, and the `before` of the first is the last.

We could implement this class more efficiently, so that operations are O(log n), by using a tree, but we'll do it the dumb way, using a list, just to make the implementation simpler, so we're sure it works right.  (It could be another lesson to implement the O(log n) methods.)

```python
class CircularSorted(list):
    "A list of items, sorted by key function, but with wrap-around so that the first item is 'after' the last."
    
    def __init__(self, items=[], key=lambda x: x):
        "Create a new list; optionally supply initial items and key function."
        self.set(items, key)
        
    def set(self, items=None, key=None):
        "Replace the items and/or the key function. Then re-sort."
        if items is not None: self[:] = items
        if key is not None: self.key = key 
        self.sort(key=self.key)
        return self # returning self allows the caller to chain operations
        
    def add(self, item):
        "Add a single new item and keep the items sorted"
        self.append(item)
        self.sort(key=self.key)
        
    def after(self, item):
        "Return the element immediately after item in sorted order. Wrap around to first element if necessary."
        i = self.index(item)
        return self[(i + 1) % len(self)]
        
    def before(self, item):
        "Return the element immediately before item in sorted order. Wrap around to last element if necessary."
        i = self.index(item)
        return self[i - 1]
```

Here are some `matplotlib` routines to let us plot diagrams:

```python
def animate_incremental_hull(points, lengths=range(2, 100)):
    "Add points to a hull one at a time; show each addition as a plot if history length is a member of lengths."
    hull = IncrementalHull()
    for p in points:
        hull.add(p)
        if len(hull.history) in lengths: plot_incremental_hull(hull)
    return hull

def plot_incremental_hull(hull):
    "Draw the points and lines of the IncrementalHull."
    plotpoints(hull.history, 'ro', range(len(hull.history))) # history as red circle; no line
    plotpoints([hull.anchor], 'ms', 'A') # anchor as magenta square
    plotpolygon(hull.hull, 'gD-') # hull points as green diamonds connected by line
    
def plotpolygon(points, style='gD-'): 
    "Like plotpoints, but wraps around to close loop.  Also does a plt.show()."
    plotpoints(points + [points[0]], style)
    plt.show()

def plotpoints(points, style='ro', labels=()): 
    "Convert Point(x, y) points into separate X, Y lists, and plot with a given style, maybe with labels."
    plt.plot([p.x for p in points], [p.y for p in points], style)
    for (p, L) in zip(points, labels):
        plt.text(p.x, p.y, '  '+str(L))
```

## Batch Convex Hull Algorithm (Graham Scan)

The batch algorithm is simpler than the incremental.  We sort the points by angle, like before (this time using the point with the lowest y coordinate as the anchor), and then we go through the points in order, throwing out the ones that make a right or straight turn.  (That is, if the last three points are ABC, and they do not make a left turn, then remove point B from the hull.)

```python
def convex_hull(points):
    "Find the hull for a batch of points, all at once."
    if len(points) <= 3: 
        return points
    # Our anchor will be the point with the lowest y coordinate.
    lowest = min(points, key=lambda p: p.y)
    # Build up the answer into hull, which is a 'regular' list
    hull = []
    for p in CircularSorted(points, key=angle_from(lowest)):
        hull.append(p)
        # if A->B->C does not form a left turn, then delete B.
        while len(hull) >= 3 and turn(hull[-3], hull[-2], hull[-1]) != 'left':
            del hull[-2]
    return hull
```

Now we can see that the incremental and batch
algorithms compute the same hull (on at least one example):

```python
def plot_convex_hull(points):
    "Use the batch convex_hull algorithm on these points, and plot the results."
    hull = convex_hull(points)
    plotpoints(points, 'r.')
    plotpolygon(hull)
    plt.show()
    print len(hull), 'of', len(points), 'points on hull'
    
plot_convex_hull(points)
animate_incremental_hull(points, [11])
print 'Are they equal?', set(convex_hull(points)) == set(animate_incremental_hull(points, ()).hull)
```

## How Many Points on the Hull?

How many points can we expect to be on the hull, as a function of the total number of points? Let's try to find out.  We'll create maps with points placed randomly for various values of *N*, the number of points.  Then we'll compute the number of points on the hull.  We'll average that over, say, 20 random samples for each value of *N*.  Then we'll plot the sesults.  

From what I've seen so far, the *N*=20 random points gave me 8 points on the hull, and *N*=2000 gave me 19.  That suggests the number of points on the hull is growing slowly with *N*. I'm going to make *N* be exponentially increasing (as powers of 2; I chose 2 because I want to get in several points between *N*=20 and *N*=2000, not just one point between them). I'll plot the exponent on the x-axis (that is, the x-axis is a log-plot) and the y-axis (the number of points on the hull) as a linear plot. 

```python
def plot_hull_sizes(base=2, exponents=range(2, 15), repeat=20):
    "Make a plot of the number of points on the hull for various number of points."
    hull_sizes = [average_hull_size(base**e, repeat) for e in exponents]
    plt.plot(exponents, hull_sizes, 'bo-', 
             label='hullsize for {}^N points (avg of {})'.format(base, repeat))
    plt.legend(loc=2)
    plt.show()
    for e, s in zip(exponents, hull_sizes):
        print('{:>6} = {}^{:<2}: {:>6.2f} hull points'.format(base**e, base, e, s))

def average_hull_size(N, repeat=20):
    "Compute the average, over given number of repeats, of the number of points on the hull of N random points."
    return average(len(convex_hull(random_points(N))) for _ in range(repeat))

# Let's try it out.  We'll set a random seed again to get the same results each time.
random.seed(123)
plot_hull_sizes()
```

This is interesting&mdash;it does look like roughly a straight line; a straight line on a log plot means log(*N*) growth.  But it looks like the number of points on the hull is levelling off, and even decreased on the final point.  Could that be a true effect?  Or is it just an accident from random sampling?  The first step to figuring that out is to repeat the experiment; I'll do it two more times, and this time with 30 repetitions for each value of *N* rather than just 20:

```python
plot_hull_sizes(repeat=30)
plot_hull_sizes(repeat=30)
```

Yes, it certainly does look like there is a real effect.  To gain more confidence, I'll repeat the test, this time with 50 repetitions at each point (so there will be less variance in the mean), and going out to $2^{16}$ = 65,536 points:

```python
plot_hull_sizes(exponents=range(2, 17), repeat=50)
```

Here it is very clear that there is a peak at $2^{12}$ = 4,096 points, with a decline after that.  *This seems very odd.*  Why would the number of points on the hull go down?  And why at that particular point?

First let's think about how adding a new point can change the number of points on the hull.  If a new point lands on the interior (or the perimeter) of the polygonal hull, then the hull remains unchanged.  If it lands on the exterior than the hull changes, and the new point becomes part of the hull, adding a point to the count.  But it is possible that other points are removed, so the total might go up or down.

Now to further explore how the number of points on the hull changes, let's think about maximal values.  When do we have the maximal number of points?  One answer would be if every point falls on the circumference of a circle, then all *N* points are part of the hull.  There are other configurations that use all *N* points, but in most random configurations, many points will be in the interior, and not part of the hull.  When do we get the minimum number of points on the hull?  Well, when almost all the points are contained within a few others.  When a large number of points are randomly sampled from a rectangle, the minimum number of points on the hull would be when there are four corner points that define the hull.  

Now we seem to be getting somewhere: maybe the decline in points on the hull has to do with the sampling process.  We are not sampling from all points on an infinite plane (if such a thing could make sense), rather we are sampling from points within a rectangle.  And even more crucially, we are sampling from *grid points* that have *integer coordinates* within that rectangle.  How many grid points?  If you go back and look at `random_points` you see we chose the ranges of random numbers to display nicely on the plot, without crowding too closely to the margins; the ranges were 25:1575 in the x-direction and 25:975 in the y-direction, for a total of (1575-25) &times; (975-25) points (about 1.5 million).  If we sampled all 1.5 million points, then we would indeed get a hull with exactly four points on it; the four corners. So it seems plausible that the decline is because we are starting to saturate the grid.  If we allowed more possible point positions (by sampling from floating point numbers rather than integers) then perhaps we wouldn't get a decline in number of points on hull.  Let's try and see.

```python
def random_points(n):
    "Redfeine random_points to select floating point coordinates, not integers."
    return [Point(random.random(), random.random()) for p in range(n)]

random.seed(float)
plot_hull_sizes()
```

Ah!  Very nice ... this really looks like a straight line (on the log graph).  In fact, so straight that it seems like we could describe it as a linear equation.  What is the slope (approximately)?  Well, the exponent ranges from 2 to 14, and the y-axis value from 3.85 to 25.05, so the slope is (25.05-3.85)/(14-2) = 1.8 (approximately).  And 1.8 times 2 is close to 3.85 (and 1.8 times 14 is close to 25.05) so let's just say the intercept is 0. 

We can modify `plot_hull_sizes` to show this line of slope 1.8 in addition to the computed average hull sizes:

```python
def plot_hull_sizes(base=2, exponents=range(2, 15), repeat=20, 
                    fn=(lambda n: 1.8 * math.log(n, 2)),
                    fn_name='1.8 log2(n)'):
    hull_sizes = [average_hull_size(base**e, repeat) for e in exponents]
    plt.plot(exponents, hull_sizes, 'bo-', 
             label='hullsize for {}^N points (avg of {})'.format(base, repeat))
    plt.plot(exponents, [fn(base**N) for N in exponents], 'rs:', label=fn_name)
    plt.legend(loc=2)
    plt.show()
    for e, s in zip(exponents, hull_sizes):
        print('{:>6} = {}^{:<2}: {:>6.2f} hull points ({:>6.2f} for {})'.format(
              base**e, base, e, s, fn(base**e), fn_name))
        
random.seed(float)
plot_hull_sizes()
```

It looks like a really good fit.  Let's try it again, this time running the exponent out to 15, and tripling the number of repetitions:

```python
plot_hull_sizes(base=2, exponents=range(2, 16), repeat=60)
```

The linear fit still looks really, really good.  I should say at this point that there is a simple method (called "*least squares*") to calculate exactly the equation of the best-fit line to the data; you don't have to just estimate like we did.  Also, there are statistical techniques to measure the amount of variation from a straight line.  But in this case the fit is so good, that statistical techniques would not add much to the story.
